# Задание 1

Определить (с обоснованием), зависимы или независимы следующие события:
(a) Несовместные события;
(b) События, образующие σ-алгебру Σ в пространстве (Ω,Σ, P);
(c) События, имеющие одинаковую вероятность?

### (а) Несовместные события

**Ответ:** Зависимы (кроме тривиального случая)

Если $A$ и $B$ несовместны ($A \cap B = \emptyset$) и $P(A) > 0$, $P(B) > 0$, то:

- $P(A \cap B) = 0$
- $P(A) \cdot P(B) > 0$
- $P(A \cap B) \ne P(A) \cdot P(B)$

Следовательно, события зависимы. Исключением является ситуация, когда хотя бы одно событие имеет нулевую вероятность.

---

### (b) События, образующие σ-алгебру $\Sigma$ в пространстве $(\Omega, \Sigma, P)$

**Ответ:** Могут быть как зависимы, так и независимы

**Обоснование:** σ-алгебра — это совокупность событий, замкнутая относительно операций. Само по себе это не накладывает ограничений на вероятности попарных пересечений.

Например:
- В σ-алгебре $\{\emptyset, A, A^c, \Omega\}$ события $A$ и $A^c$ зависимы
- Можно построить σ-алгебру с независимыми событиями

---

### (c) События, имеющие одинаковую вероятность

**Ответ:** Могут быть как зависимы, так и независимы

**Обоснование:** Равенство вероятностей $P(A) = P(B)$ не даёт информации о $P(A \cap B)$.

**Примеры:**
- **Независимые:** $P(A) = P(B) = 0.5$, $P(A \cap B) = 0.25$
- **Зависимые:** $P(A) = P(B) = 0.5$, $P(A \cap B) = 0.4$

# Задание 4

## Условие задачи

Два парохода подходят к одному причалу:
- Независимо друг от друга
- Равновозможно в течение суток
- Время стоянки: первый — 1 час, второй — 2 часа

**Найти:** Вероятность того, что одному придётся ожидать освобождения причала

---

## Решение

### Шаг 1. Моделирование задачи

Введём обозначения:
- $x$ — время прихода первого парохода (равномерно на [0,24])
- $y$ — время прихода второго парохода (равномерно на [0,24])
- Первый пароход стоит 1 час: занимает интервал $[x, x+1]$
- Второй пароход стоит 2 часа: занимает интервал $[y, y+2]$

**Ожидание происходит**, если интервалы стоянки пересекаются.

---

### Шаг 2. Применение формулы пересечения интервалов

Для двух независимых случайных величин, равномерно распределённых на $[0,T]$, с длительностями пребывания $t_1$ и $t_2$ ($t_1 + t_2 \leq T$), вероятность пересечения интервалов равна:

$$
P = 1 - \frac{(T - t_1)(T - t_2)}{T^2}
$$

---

### Шаг 3. Подстановка значений

- $T = 24$ часов (сутки)
- $t_1 = 1$ час (первый пароход)
- $t_2 = 2$ часа (второй пароход)

Проверяем условие: $t_1 + t_2 = 3 \leq 24$ — выполняется.

Вычисляем:

$$
P = 1 - \frac{(24 - 1)(24 - 2)}{24^2} = 1 - \frac{23 \cdot 22}{576} = 1 - \frac{506}{576} = \frac{70}{576}
$$

---

### Шаг 4. Упрощение ответа

$$
P = \frac{70}{576} = \frac{35}{288}
$$

---

## Ответ

Вероятность того, что одному из пароходов придётся ожидать освобождения причала:

$$
\frac{35}{288}
$$

Это примерно $0.1215$

# Задание 5

## Условие
Самолет состоит из трёх частей с разными условиями поражения:
- **Часть 1**: 1 попадание
- **Часть 2**: 2 попадания  
- **Часть 3**: 3 попадания

Вероятности попадания в части: $p_1$, $p_2$, $p_3$ ($p_1 + p_2 + p_3 = 1$)

**Найти:** $P(A|m)$ для $m = 1, 2, 3, 4$

---

## Решение

### Обозначения
- $x_1$ — число попаданий в часть 1
- $x_2$ — число попаданий в часть 2  
- $x_3$ — число попаданий в часть 3

**Самолет поражён, если:**
- $x_1 \geq 1$, или
- $x_2 \geq 2$, или  
- $x_3 \geq 3$

---

### Случай m = 1
При одном снаряде поражение возможно только при попадании в часть 1:

$$P(A|1) = P(x_1 = 1) = p_1$$

---

### Случай m = 2
**Поражения НЕТ, если:**
- $x_1 = 0$ и $x_2 \leq 1$ и $x_3 \leq 2$

**Возможные комбинации:**
1. $(0, 0, 2)$: вероятность $p_3^2$
2. $(0, 1, 1)$: вероятность $2p_2p_3$

$$P(\bar{A}|2) = p_3^2 + 2p_2p_3 = p_3(p_3 + 2p_2)$$

$$P(A|2) = 1 - p_3(p_3 + 2p_2)$$

---

### Случай m = 3
**Поражения НЕТ, если:**
- $x_1 = 0$ и $x_2 \leq 1$ и $x_3 \leq 2$

**Единственная возможная комбинация:**
- $(0, 1, 2)$: вероятность $3p_2p_3^2$

$$P(\bar{A}|3) = 3p_2p_3^2$$

$$P(A|3) = 1 - 3p_2p_3^2$$

---

### Случай m = 4
При четырёх снарядах **любая** комбинация даёт поражение:
- Если $x_1 \geq 1$ — поражение
- Если $x_1 = 0$, то все 4 снаряда распределены между частями 2 и 3
- Но тогда либо $x_2 \geq 2$, либо $x_3 \geq 3$

$$P(A|4) = 1$$

---

## Итоговый ответ

$$
P(A|1) = p_1
$$

$$
P(A|2) = 1 - p_3(p_3 + 2p_2)
$$

$$
P(A|3) = 1 - 3p_2p_3^2
$$

$$
P(A|4) = 1
$$

# Задание 1
Пусть

$$
f_{\xi}(x, y) = \frac{e^{-2|y|}}{\pi(1 + x^2)}
$$

Является ли данная функция плотностью распределения случайного
вектора?

### Условия плотности распределения

1. Функция неотрицательна:
   $$
   f(x, y) \ge 0 \quad \forall\, x, y
   $$

2. Интеграл по всей плоскости равен единице:
   $$
   \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f(x, y)\,dx\,dy = 1
   $$

Функция положительна, так как экспонента и знаменатель положительны.
Интеграл распадается на произведение двух независимых интегралов:
$$
\left( \int_{-\infty}^{\infty} \frac{dx}{\pi(1 + x^2)} \right)
\cdot
\left( \int_{-\infty}^{\infty} e^{-2|y|}\,dy \right)
= 1 \cdot 1 = 1
$$

то есть функция действительно является плотностью распределения.

# Задание 2
Совместное распределение случайных величин ξ и η задано следующей
таблицей:

| $\xi \backslash \eta$ | $-1$ | $0$ | $1$ |
|:----------------------:|:----:|:---:|:---:|

---

### (a)
Найти **маргинальные распределения** $\xi$ и $\eta$:
$$
P_\xi(x_i) = \sum_j P(x_i, y_j), \qquad
P_\eta(y_j) = \sum_i P(x_i, y_j)
$$

---

### (b)
Вычислить:
- математические ожидания $E[\xi]$, $E[\eta]$;  
- ковариационную и корреляционную матрицы вектора $(\xi, \eta)$:
$$
\operatorname{Cov}(\xi,\eta) = E[\xi \eta] - E[\xi]E[\eta], \qquad
\rho_{\xi\eta} = \frac{\operatorname{Cov}(\xi,\eta)}{\sigma_\xi \sigma_\eta}
$$

### (c)
Исследовать случайные величины $\xi$ и $\eta$ на:
- **независимость**  
  $$
  P(x_i, y_j) = P_\xi(x_i) P_\eta(y_j)
  $$
- **некоррелированность**  
  $$
  \operatorname{Cov}(\xi,\eta) = 0
  $$


## **(a) Маргинальные распределения**

### **Для $\xi$**

Суммируем вероятности по строкам:

$$
P(\xi = -1) = \frac{1}{8} + \frac{1}{12} + \frac{7}{24} = \frac{3}{24} + \frac{2}{24} + \frac{7}{24} = \frac{12}{24} = 0.5
$$

$$
P(\xi = 1) = \frac{1}{3} + \frac{1}{6} + 0 = \frac{4}{12} + \frac{2}{12} = \frac{6}{12} = 0.5
$$

**Маргинальное распределение $\xi$:**

$$
P(\xi = -1) = 0.5, \qquad P(\xi = 1) = 0.5
$$

---

### **Для $\eta$**

Суммируем по столбцам:

$$
P(\eta = -1) = \frac{1}{8} + \frac{1}{3} = \frac{3}{24} + \frac{8}{24} = \frac{11}{24}
$$

$$
P(\eta = 0) = \frac{1}{12} + \frac{1}{6} = \frac{2}{24} + \frac{4}{24} = \frac{6}{24} = \frac{1}{4}
$$

$$
P(\eta = 1) = \frac{7}{24} + 0 = \frac{7}{24}
$$

**Маргинальное распределение $\eta$:**

$$
P(\eta = -1) = \frac{11}{24}, \quad P(\eta = 0) = \frac{6}{24}, \quad P(\eta = 1) = \frac{7}{24}
$$

---

## **(b) Математические ожидания, ковариация, корреляция**

### **Математическое ожидание $\xi$**

$$
\mathbb{E}[\xi] = (-1) \cdot 0.5 + 1 \cdot 0.5 = 0
$$

---

### **Математическое ожидание $\eta$**

$$
\mathbb{E}[\eta] = (-1) \cdot \frac{11}{24} + 0 \cdot \frac{6}{24} + 1 \cdot \frac{7}{24} = -\frac{11}{24} + \frac{7}{24} = -\frac{4}{24} = -\frac{1}{6}
$$

---

### **Смешанное ожидание $\mathbb{E}[\xi\eta]$**

$$
\begin{aligned}
\mathbb{E}[\xi\eta] &= (-1)(-1) \cdot \frac{1}{8} + (-1)(0) \cdot \frac{1}{12} + (-1)(1) \cdot \frac{7}{24} \\
&\quad + (1)(-1) \cdot \frac{1}{3} + (1)(0) \cdot \frac{1}{6} + (1)(1) \cdot 0 \\
&= \frac{1}{8} + 0 - \frac{7}{24} - \frac{1}{3} = \frac{3}{24} - \frac{7}{24} - \frac{8}{24} = -\frac{12}{24} = -\frac{1}{2}
\end{aligned}
$$

---

### **Ковариация**

$$
\mathrm{Cov}(\xi,\eta) = \mathbb{E}[\xi\eta] - \mathbb{E}[\xi]\mathbb{E}[\eta] = -\frac{1}{2} - 0 = -\frac{1}{2}
$$

Отрицательная ковариация: при росте $\xi$ величина $\eta$ обычно уменьшается.

---

### **Дисперсии**

Для $\xi$:

$$
\mathbb{D}[\xi] = (-1 - 0)^2 \cdot 0.5 + (1 - 0)^2 \cdot 0.5 = 1
$$

Для $\eta$:

$$
\mathbb{D}[\eta] = \left(-1 + \frac{1}{6}\right)^2 \cdot \frac{11}{24} + \left(\frac{1}{6}\right)^2 \cdot \frac{6}{24} + \left(1 + \frac{1}{6}\right)^2 \cdot \frac{7}{24} = \frac{49}{144} \approx 0.340
$$

---

### **Корреляция**

$$
\rho_{\xi\eta} = \frac{\mathrm{Cov}(\xi,\eta)}{\sqrt{\mathbb{D}[\xi] \cdot \mathbb{D}[\eta]}} = \frac{-\frac{1}{2}}{\sqrt{1 \cdot 0.340}} \approx -0.588
$$

Наблюдается средне-сильная отрицательная связь: $\xi$ и $\eta$ «движутся в разные стороны».

---

## **(c) Независимость и некоррелированность**

### **Проверка независимости**

$$
P(x_i, y_j) \stackrel{?}{=} P_\xi(x_i) P_\eta(y_j)
$$

Проверка на примере:

$$
P(\xi = -1, \eta = -1) = \frac{1}{8}
$$

$$
P(\xi = -1) P(\eta = -1) = 0.5 \cdot \frac{11}{24} = \frac{11}{48}
$$

$$
\frac{1}{8} \ne \frac{11}{48}
$$

**Следовательно:** $\xi$ и $\eta$ — зависимы.

---

### **Проверка некоррелированности**

$$
\mathrm{Cov}(\xi, \eta) = -\frac{1}{2} \ne 0
$$

**Следовательно:** $\xi$ и $\eta$ — коррелированы.

E[\phi_1]=2\frac1{16}+3\frac2{16}+4\frac3{16}+5\frac4{16}+6\frac3{16}+7\frac2{16}+8\frac1{16}=5
$

### **Ожидание $E[\phi_2]$**

$
E[\phi_2]=0\cdot \frac4{16} + 1\cdot \frac{12}{16}
= \frac{12}{16}
$

---

### **Ожидание произведения**

$
E[\phi_1\phi_2]=\sum_{k} k P(\phi_1=k,\phi_2=1)
$

$
E[\phi_1\phi_2]
=\frac{1}{16}(2\cdot1+3\cdot2+4\cdot2+5\cdot1+6\cdot2+7\cdot2+8\cdot1)
$

$
E[\phi_1\phi_2]=\frac{56}{16}=3.5
$

---

### Ковариационная матрица


#### $D[\phi_1]$:
$$
E[\phi_1^2] = 4\cdot\frac{1}{16} + 9\cdot\frac{2}{16} + 16\cdot\frac{3}{16} + 25\cdot\frac{4}{16} + 36\cdot\frac{3}{16} + 49\cdot\frac{2}{16} + 64\cdot\frac{1}{16}
$$

E[\phi_1^2] = \frac{4+18+48+100+108+98+64}{16} = \frac{440}{16} = 27.5
$$

$$
D[\phi_1] = E[\phi_1^2] - (E[\phi_1])^2 = 27.5 - 25 = 2.5 = \frac{5}{2}
$$

#### $D[\phi_2]$:
$$
E[\phi_2^2] = 0^2\cdot\frac14 + 1^2\cdot\frac34 = \frac34
$$

$$
D[\phi_2] = E[\phi_2^2] - (E[\phi_2])^2 = \frac34 - \left(\frac34\right)^2 = \frac{12-9}{16} = \frac{3}{16}
$$

### Ковариация

$$
\text{cov}(\phi_1,\phi_2) = E[\phi_1\phi_2] - E[\phi_1]E[\phi_2] = 3.5 - 5\cdot\frac34 = -0.25 = -\frac14
$$

K = \begin{pmatrix}
D[\phi_1] & \text{cov}(\phi_1,\phi_2) \\
\text{cov}(\phi_1,\phi_2) & D[\phi_2]
\end{pmatrix}

\frac{5}{2} & -\frac14 \\

\end{pmatrix}
$$

### Корреляционная матрица

#### Коэффициент корреляции:

\rho = \frac{\text{cov}(\phi_1,\phi_2)}{\sqrt{D[\phi_1] \cdot D[\phi_2]}} = \frac{-\frac14}{\sqrt{\frac{5}{2} \cdot \frac{3}{16}}}
$$

$$
\rho = \frac{-\frac14}{\sqrt{\frac{15}{32}}} = \frac{-\frac14}{\frac{\sqrt{15}}{\sqrt{32}}} = -\frac{\sqrt{2}}{\sqrt{15}} = -\sqrt{\frac{2}{15}}
$$

#### Корреляционная матрица:
$$
R = \begin{pmatrix}
1 & -\sqrt{\frac{2}{15}} \\
-\sqrt{\frac{2}{15}} & 1
\end{pmatrix}
$$

### (d) Независимость и некоррелированность

### **Независимость**
$
P(\phi_1 = a, \phi_2 = b) = P(\phi_1 = a) \cdot P(\phi_2 = b) \quad \text{для всех } a,b
$

**Некоррелированность случайных величин**:
$
\text{cov}(\phi_1, \phi_2) = 0 \quad \text{или} \quad \rho_{\phi_1\phi_2} = 0
$


- Независимость $\Rightarrow$ некоррелированность
- Некоррелированность $\not\Rightarrow$ независимость


**Проверим ячейку $\phi_1 = 2$, $\phi_2 = 1$:**

Совместная вероятность:
$
P(\phi_1 = 2, \phi_2 = 1) = \frac{1}{16}
$

Произведение маргинальных вероятностей:
$
P(\phi_1 = 2) \cdot P(\phi_2 = 1) = \frac{1}{16} \cdot \frac{3}{4} = \frac{3}{64}
$

Сравниваем:
$
\frac{1}{16} = \frac{4}{64} \neq \frac{3}{64}
$

**Вывод**: Условие независимости нарушено.


### **Проверка некоррелированности**

Из предыдущих вычислений:

Ковариация:
$
\text{cov}(\phi_1, \phi_2) = -\frac{1}{4} \neq 0
$

Коэффициент корреляции:
$
\rho_{\phi_1\phi_2} = -\sqrt{\frac{2}{15}} \approx -0.365 \neq 0
$

**Вывод**: Величины коррелированы.


**Интерпретация**:
- Отрицательная корреляция ($\rho \approx -0.365$) означает, что с ростом суммы $\phi_1$ вероятность выполнения условия делимости $\phi_2 = 1$ уменьшается
- Например: при $\phi_1 = 7,8$ в большинстве случаев $\phi_2 = 0$, а при $\phi_1 = 2,3,4,6$ всегда $\phi_2 = 1$


# Задание 4

### Дано:
Случайная величина $\xi \sim U(-\pi, \pi)$ с плотностью распределения:

$$
f_\xi(x) = \frac{1}{2\pi}, \quad x \in [-\pi, \pi]
$$

Определены функции:

$$
\eta_1 = \cos\xi, \quad \eta_2 = \sin\xi
$$

Рассматриваем вектор $(\xi, \eta_1, \eta_2)$.

---

## **(a) Математическое ожидание, ковариационная и корреляционная матрицы**

### **1. Математические ожидания**

Для равномерного распределения на симметричном интервале:

$$
\mathbb{E}[\xi] = \int_{-\pi}^{\pi} x \cdot \frac{1}{2\pi}  dx = 0
$$

Для $\eta_1$:

$$
\mathbb{E}[\eta_1] = \mathbb{E}[\cos\xi] = \int_{-\pi}^{\pi} \cos x \cdot \frac{1}{2\pi}  dx = \frac{1}{2\pi} [\sin x]_{-\pi}^{\pi} = 0
$$

Для $\eta_2$:

$$
\mathbb{E}[\eta_2] = \mathbb{E}[\sin\xi] = \int_{-\pi}^{\pi} \sin x \cdot \frac{1}{2\pi}  dx = \frac{1}{2\pi} [-\cos x]_{-\pi}^{\pi} = 0
$$

**Итог:**

$$
\mathbb{E}[\xi] = 0, \quad \mathbb{E}[\eta_1] = 0, \quad \mathbb{E}[\eta_2] = 0
$$

---

### **2. Дисперсии и ковариации**

**Дисперсия $\xi$:**

$$
\mathbb{D}[\xi] = \mathbb{E}[\xi^2] = \int_{-\pi}^{\pi} x^2 \cdot \frac{1}{2\pi}  dx = \frac{1}{2\pi} \left[ \frac{x^3}{3} \right]_{-\pi}^{\pi} = \frac{1}{2\pi} \cdot \frac{2\pi^3}{3} = \frac{\pi^2}{3}
$$

**Дисперсия $\eta_1$:**

$$
\mathbb{D}[\eta_1] = \mathbb{E}[\cos^2\xi] = \int_{-\pi}^{\pi} \cos^2 x \cdot \frac{1}{2\pi}  dx
$$

Используем тригонометрическое тождество $\cos^2 x = \frac{1 + \cos 2x}{2}$:

$$
\mathbb{D}[\eta_1] = \frac{1}{2\pi} \int_{-\pi}^{\pi} \frac{1 + \cos 2x}{2}  dx = \frac{1}{4\pi} \left[ x + \frac{\sin 2x}{2} \right]_{-\pi}^{\pi} = \frac{1}{4\pi} \cdot 2\pi = \frac{1}{2}
$$

**Дисперсия $\eta_2$:**

$$
\mathbb{D}[\eta_2] = \mathbb{E}[\sin^2\xi] = \int_{-\pi}^{\pi} \sin^2 x \cdot \frac{1}{2\pi}  dx
$$

Используем $\sin^2 x = \frac{1 - \cos 2x}{2}$:

$$
\mathbb{D}[\eta_2] = \frac{1}{2\pi} \int_{-\pi}^{\pi} \frac{1 - \cos 2x}{2}  dx = \frac{1}{4\pi} \left[ x - \frac{\sin 2x}{2} \right]_{-\pi}^{\pi} = \frac{1}{4\pi} \cdot 2\pi = \frac{1}{2}
$$

---

### **3. Ковариации**

**Ковариация $\xi$ и $\eta_1$:**

$$
\mathrm{cov}(\xi, \eta_1) = \mathbb{E}[\xi \cos\xi] = \int_{-\pi}^{\pi} x \cos x \cdot \frac{1}{2\pi}  dx = 0
$$

Подынтегральная функция $x\cos x$ — нечётная, интеграл по симметричному промежутку равен 0.

**Ковариация $\xi$ и $\eta_2$:**

$$
\mathrm{cov}(\xi, \eta_2) = \mathbb{E}[\xi \sin\xi] = \int_{-\pi}^{\pi} x \sin x \cdot \frac{1}{2\pi}  dx
$$

Функция $x\sin x$ — чётная, вычисляем интеграл:

$$
\int x \sin x  dx = -x\cos x + \sin x + C
$$

$$
\mathbb{E}[\xi \sin\xi] = \frac{1}{2\pi} \left[ -x\cos x + \sin x \right]_{-\pi}^{\pi} = \frac{1}{2\pi} \cdot 2\pi = 1
$$

**Ковариация $\eta_1$ и $\eta_2$:**

$$
\mathrm{cov}(\eta_1, \eta_2) = \mathbb{E}[\cos\xi \sin\xi] = \mathbb{E}\left[ \frac{\sin 2\xi}{2} \right] = \frac{1}{2} \int_{-\pi}^{\pi} \sin 2x \cdot \frac{1}{2\pi}  dx = 0
$$

Интеграл от синуса на полном периоде равен 0.

---

### **4. Ковариационная матрица**

$$
K = \begin{pmatrix}
\mathbb{D}[\xi] & \mathrm{cov}(\xi, \eta_1) & \mathrm{cov}(\xi, \eta_2) \\
\mathrm{cov}(\eta_1, \xi) & \mathbb{D}[\eta_1] & \mathrm{cov}(\eta_1, \eta_2) \\
\mathrm{cov}(\eta_2, \xi) & \mathrm{cov}(\eta_2, \eta_1) & \mathbb{D}[\eta_2]
\end{pmatrix}
= \begin{pmatrix}
\frac{\pi^2}{3} & 0 & 1 \\
0 & \frac{1}{2} & 0 \\
1 & 0 & \frac{1}{2}
\end{pmatrix}
$$

---

### **5. Корреляционная матрица**

Корреляция вычисляется по формуле:

$$
r_{ij} = \frac{\mathrm{cov}(X_i, X_j)}{\sqrt{\mathbb{D}[X_i] \cdot \mathbb{D}[X_j]}}
$$

- $r_{\xi,\eta_1} = 0$
- $r_{\xi,\eta_2} = \dfrac{1}{\sqrt{\frac{\pi^2}{3} \cdot \frac{1}{2}}} = \dfrac{\sqrt{6}}{\pi}$
- $r_{\eta_1,\eta_2} = 0$

$$
R = \begin{pmatrix}
1 & 0 & \frac{\sqrt{6}}{\pi} \\
0 & 1 & 0 \\
\frac{\sqrt{6}}{\pi} & 0 & 1
\end{pmatrix}
$$

---

## **(b) Исследование на независимость и некоррелированность**

### **1. $\xi$ и $\eta_1$**

- Ковариация: $\mathrm{cov}(\xi, \eta_1) = 0$ → **некоррелированы**
- Однако $\eta_1 = \cos\xi$ — функциональная зависимость → **зависимы**

### **2. $\xi$ и $\eta_2$**

- Ковариация: $\mathrm{cov}(\xi, \eta_2) = 1 \neq 0$ → **коррелированы** → **зависимы**

### **3. $\eta_1$ и $\eta_2$**

- Ковариация: $\mathrm{cov}(\eta_1, \eta_2) = 0$ → **некоррелированы**
- Но $\eta_1^2 + \eta_2^2 = \cos^2\xi + \sin^2\xi = 1$ — детерминированная связь → **зависимы**

---

## **Выводы:**

- Все компоненты вектора $(\xi, \eta_1, \eta_2)$ попарно зависимы
- Пары $(\xi, \eta_1)$ и $(\eta_1, \eta_2)$ — некоррелированы, но зависимы
- Пара $(\xi, \eta_2)$ — коррелирована и зависима
- Некоррелированность не влечёт независимость для нелинейно связанных величин

# Задание 5
## Найти плотность распределения суммы двух независимых случайных величин

### Дано:
- $\xi \sim Exp(2)$ - показательное распределение с параметром $\lambda = 2$
- $\eta \sim U(0,1)$ - равномерное распределение на $[0,1]$
- $\xi$ и $\eta$ - независимы
- Найти плотность распределения $\zeta = \xi + \eta$

---

## Решение

### Шаг 1: Запишем плотности исходных распределений

**Плотность показательного распределения:**

$$
f_\xi(x) = \begin{cases}
2e^{-2x}, & x \geq 0 \\
0, & x < 0
\end{cases}
$$

**Плотность равномерного распределения:**

$$
f_\eta(y) = \begin{cases}
1, & 0 \leq y \leq 1 \\
0, & \text{иначе}
\end{cases}
$$

---

### Шаг 2: Формула свёртки

Для суммы независимых случайных величин плотность находится по формуле свёртки:

$$
f_\zeta(z) = \int_{-\infty}^{\infty} f_\xi(x) f_\eta(z-x)  dx
$$

---

### Шаг 3: Проанализируем области, где подынтегральное выражение не равно нулю

Учитываем носители плотностей:
- $f_\xi(x) \neq 0$ при $x \geq 0$
- $f_\eta(z-x) \neq 0$ при $0 \leq z-x \leq 1$, т.е. $z-1 \leq x \leq z$

**Область интегрирования определяется системой:**

$$
\begin{cases}
x \geq 0 \\
z-1 \leq x \leq z
\end{cases}
$$

---

### Шаг 4: Рассмотрим различные случаи для $z$

#### **Случай 1: $z < 0$**
Области $x \geq 0$ и $x \leq z$ не пересекаются, так как $z < 0$.

$$
f_\zeta(z) = 0
$$

#### **Случай 2: $0 \leq z \leq 1$**
Область интегрирования: от $x = 0$ до $x = z$

$$
f_\zeta(z) = \int_{0}^{z} 2e^{-2x} \cdot 1  dx = 2 \int_{0}^{z} e^{-2x}  dx
$$

$$
= 2 \left[ -\frac{1}{2}e^{-2x} \right]_{0}^{z} = -[e^{-2z} - 1] = 1 - e^{-2z}
$$

#### **Случай 3: $z > 1$**
Область интегрирования: от $x = z-1$ до $x = z$

$$
f_\zeta(z) = \int_{z-1}^{z} 2e^{-2x} \cdot 1  dx = 2 \int_{z-1}^{z} e^{-2x}  dx
$$

$$
= 2 \left[ -\frac{1}{2}e^{-2x} \right]_{z-1}^{z} = -[e^{-2z} - e^{-2(z-1)}]
$$

$$
= e^{-2(z-1)} - e^{-2z} = e^{-2z}(e^{2} - 1)
$$

---

### Шаг 5: Итоговая плотность распределения

$$
f_\zeta(z) = \begin{cases}
0, & z < 0 \\
1 - e^{-2z}, & 0 \leq z \leq 1 \\
e^{-2z}(e^{2} - 1), & z > 1
\end{cases}
$$

---

## Проверка

**Убедимся, что плотность нормирована:**

$$
\int_{0}^{1} (1 - e^{-2z})  dz + \int_{1}^{\infty} e^{-2z}(e^{2} - 1)  dz
$$

$$
= \left[ z + \frac{1}{2}e^{-2z} \right]_{0}^{1} + (e^{2} - 1) \int_{1}^{\infty} e^{-2z}  dz
$$

$$
= \left(1 + \frac{1}{2}e^{-2} - \frac{1}{2}\right) + (e^{2} - 1) \left[ -\frac{1}{2}e^{-2z} \right]_{1}^{\infty}
$$

$$
= \left(\frac{1}{2} + \frac{1}{2}e^{-2}\right) + (e^{2} - 1) \cdot \frac{1}{2}e^{-2}
$$

$$
= \frac{1}{2} + \frac{1}{2}e^{-2} + \frac{1}{2} - \frac{1}{2}e^{-2} = 1
$$

Плотность нормирована правильно.

---

## Геометрическая интерпретация

- При $0 \leq z \leq 1$: "хвост" экспоненты успевает полностью проинтегрироваться по всей длине равномерного распределения
- При $z > 1$: интегрируем только по последнему участку экспоненты длиной 1
- При $z < 0$: сумма не может быть отрицательной, так как обе величины неотрицательные
